```python
import matplotlib.pyplot as plt 

#nomes dos dados
labels = 'Meninos', 'Meninas'
#tamanho de cada label no gráfico
sizes = [40, 60]

cor = ['#87CEEB','#6A5ACD']

#cria a representação, área de plotagem
fig1, ax1 = plt.subplots()

#título do gráfico
ax1.set_title('Quantidade que almoçou por sexo')

'''cria o gráfico
autopct = permite exibir o valor percentual usando a formatação de string 
shadow = sombra
startangle = o angulo que vai partir os dados
'''
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False,    startangle=270, colors=cor)

#assim o gráfico fica em círculo
ax1.axis('equal')

#mostra o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 6), 
                       subplot_kw=dict(aspect="equal"))

#dados para compor o gráfico
sexo = ['Masculino', 'Feminino']
#qtde de cada dado
dados = [324, 376]
cor = ['#87CEEB','#6A5ACD']

#adiciona as porcentagens
'''
func(pct, allvals) -> calcula % e mostra legenda interna
np.sum(allvals) -> soma todos os valores
'''
def func(pct, allvals):
    #calc %
    absolute = int(pct/100.*np.sum(allvals))
    #faz legenda pra aparecer % e qtde bruta de pessoas
    return '{:.1f}%\n{:d} pessoas'.format(pct, absolute)


'''
textprops=dict(color='w') -> define cor do texto
autopct -> permite exibir o valor percentual usando a formatação de string 
lambda pct: func(pct, dados) -> para cada valor dos dados, acontece o autopct
'''

#criando gráfico de pizza e colocando a funcao na legenda interna
wedges, texts, autotexts = ax.pie(dados, 
                                  autopct=lambda pct: func(pct, dados), 
                                  colors=cor)

'''
caixa de legenda externa -> wedges, sexo
loc -> localizacao
bbox_to_anchor -> onde vai ancorar o box
'''
ax.legend(wedges, sexo, 
          title='Sexo', 
          loc='center left', 
          bbox_to_anchor=(1, 0, 0.5, 1))

#size -> tamanho do texto (de dentro do gráfico)
#weight -> peso da fonte como bold (negrito)
plt.setp(autotexts, size = 10, 
         weight= 'bold')

#título do gráfico
ax.set_title('Quantidade que almoçou por sexo')

plt.show()
```
