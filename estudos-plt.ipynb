```python
import matplotlib.pyplot as plt 

#nomes dos dados
labels = 'Meninos', 'Meninas'
#tamanho de cada label no gráfico
sizes = [40, 60]

cor = ['#87CEEB','#6A5ACD']

#cria a representação, área de plotagem
fig1, ax1 = plt.subplots()

#título do gráfico
ax1.set_title('Quantidade que almoçou por sexo')

'''cria o gráfico
autopct = 
shadow = sombra
startangle =
'''
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False,    startangle=90, colors=cor)

#assim o gráfico fica em círculo
ax1.axis('equal')

#mostra o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

sexo = ['Masculino', 'Feminino']

dados = [324, 376]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return '{:.1f}%\n{:d} pessoas'.format(pct, absolute)

wedges, texts, autotexts = ax.pie(dados, autopct=lambda pct: func(pct, dados), textprops=dict(color='w'))

ax.legend(wedges, sexo, title='Sexo', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight= 'bold')

ax.set_title('Quantidade que almoçou por sexo')

plt.show()
```
